## Data Analysis Using Python - Blinkint Analysis

Blinkit is an Indian quick-commerce company owned by Eternal Limited. It was founded in December 2013 and is based in Gurgaon.

Customers of the company use a mobile application to order groceries and essentials online. Blinkit's employees then secure the items from their warehouse and deliver the items to the consumer within 10 minutes.

**Business Requirements:**
To conduct a comprehensive analysis of Blinkit's sales performance, customer satisfaction, and inventory distribution to
identify key insights and opportunities for optimization using various KPIs and visualizations in Power BI.

**1)	KPI's Requirements**

1. Total Sales: The overall revenue generated from all items sold.
2. Average Sales: The average revenue per sale.
3. Number of Items: The total count of different items sold.
4. Average Rating: The average customer rating for items sold.

## Business Requirements

### KPI's Requirements

```python
#Total Sales
total_sales = df['Sales'].sum()

#Average Sales
avg_sales = df['Sales'].mean()

#Number of Items sold
no_of_items_sold = df['Sales'].count()

#Average Ratings
avg_ratings = df['Rating'].mean()

#Display Output
print(f"Total Sales: ${total_sales:,.1f}")
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No Of Items Sold: {no_of_items_sold:,.1f}")
print(f"Average Ratings: {avg_ratings:,.1f}")
```

### Charts Requirements

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
```

#### **Total Sales by Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

plt.pie(sales_by_fat, labels = sales_by_fat.index,
       autopct= '%1f',
       startangle=90,
       colors=colors)

plt.title('Total Sales by Fat Content')
plt.axis('equal')
plt.show()
```

#### **Average Sales by Fat Content**


```python
avg_sales_by_fat_content = df.groupby('Item Fat Content')['Sales'].mean()

plt.pie(avg_sales_by_fat_content,labels = avg_sales_by_fat_content.index,
       autopct='%1f',
       startangle=90,
       colors= colors)
plt.axis('equal')
plt.title('Average Sales by Fat Content')
plt.show()
```

#### **Number of Items by Fat Content**


```python
no_of_items_sold_by_fat_content = df.groupby('Item Fat Content')['Sales'].count()
plt.pie(no_of_items_sold_by_fat_content, labels = no_of_items_sold_by_fat_content.index,
       autopct='%1f',
       startangle=90,
       colors=colors)
plt.axis('equal')
plt.title('Number of Items Sold by Fat Content')
plt.show()
```

#### **Average Rating by Fat Content**


```python
avg_rating_by_fat_content = df.groupby('Item Fat Content')['Rating'].mean()

plt.pie(avg_rating_by_fat_content, labels= avg_rating_by_fat_content.index,
       autopct='%1f',
       startangle=90,
       colors=colors)
plt.axis('equal')
plt.title('Average Rating by Fat Content')
plt.show()
```

#### **Total Sales by Item Type**

```python
sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize = (10,6))
bars = plt.bar(sales_by_item_type.index, sales_by_item_type.values, color= colors)

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height():,.0f}', ha='center',va='bottom', fontsize='8')

plt.tight_layout()
plt.show()
```

#### **Average sales by Item Type**

```python
avg_sales_by_item_type =  df.groupby('Item Type')['Sales'].mean().sort_values(ascending=False)

plt.figure(figsize=(10,6))
bars= plt.bar(avg_sales_by_item_type.index, avg_sales_by_item_type.values, color = colors)

plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Average Sales')
plt.title('Average Sales by Item Type')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            f'{bar.get_height():,.0f}', ha = 'center', va='bottom', fontsize = 8)

plt.tight_layout()
plt.show()


```

#### **Fat Content by Outlet And Total Sales:**

Compare total sales across different outlets segmented by fat content. Using Stacked Column Chart

```python
fat_content_by_outlet_for_total_sales = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
fat_content_by_outlet_for_total_sales = fat_content_by_outlet_for_total_sales[['Low Fat', 'Regular']]

ax = fat_content_by_outlet_for_total_sales.plot(kind= 'bar', figsize = (8,5), title = 'Fat Content By Outlet and Total Sales',color=colors)

plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet And Average Sales:**

```python
fat_content_by_outlet_for_average_sales = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].mean().unstack()
fat_content_by_outlet_for_average_sales = fat_content_by_outlet_for_average_sales[['Low Fat', 'Regular']]

ax = fat_content_by_outlet_for_average_sales.plot(kind = 'bar', figsize = (10,5), title = 'Fat Content by Outlet and Average Sales', color = colors)

plt.xlabel('Outlet Location Tier')
plt.ylabel('Average Sales')
plt.legend(title= 'Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet And No of Items Sold:**

```python
fat_content_by_outlet_for_no_of_items_sold = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].count().unstack()
fat_content_by_outlet_for_no_of_items_sold = fat_content_by_outlet_for_no_of_items_sold[['Low Fat', 'Regular']]

ax = fat_content_by_outlet_for_no_of_items_sold.plot(kind = 'bar', figsize = (10,5), title = 'Fat Content by Outlet and No of Items sold', color = colors)

plt.xlabel('Outlet Location Tier')
plt.ylabel('No of Items Sold')
plt.legend(title= 'Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Fat Content by Outlet and Average Rating:**

```python
fat_content_by_outlet_for_avg_rating = df.groupby(['Outlet Location Type','Item Fat Content'])['Rating'].mean().unstack()
fat_content_by_outlet_for_avg_rating = fat_content_by_outlet_for_avg_rating[['Low Fat', 'Regular']]

ax = fat_content_by_outlet_for_avg_rating.plot(kind = 'bar', figsize = (10,5), title = 'Fat Content by Outlet and Average Rating', color = colors)

plt.xlabel('Outlet Location Tier')
plt.ylabel('Average Rating')
plt.legend(title= 'Item Fat Content')
plt.tight_layout()
plt.show()
```

#### **Total Sales by Outlet Establishment:**
Evaluate how the age or type of outlet establishment influences total sales.

```python
total_sales_by_outlet_establishment = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

plt.figure(figsize=(9,5))
plt.plot(total_sales_by_outlet_establishment.index, total_sales_by_outlet_establishment.values, marker = 'o', linestyle = '--', color= 'gold')

plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')

for x,y in zip(total_sales_by_outlet_establishment.index, total_sales_by_outlet_establishment.values):
    plt.text(x,y, f'{y:,.0f}', ha='center', va='bottom', fontsize = 8)

plt.tight_layout()
plt.show()
```

#### **Sales by Outlet Size:**
Analyze the correlation between outlet size and total sales.

```python
sales_by_outlet_size = df.groupby('Outlet Size')['Sales'].sum()

plt.figure(figsize=(4,4))
plt.pie(sales_by_outlet_size, labels = sales_by_outlet_size.index, 
       autopct='%1f',
       startangle=90,
       colors=colors)

plt.title('Total Sales by Outlet Size')
plt.tight_layout()
plt.show()

```

#### **Sales by Outlet Location:**
Assess the geographic distribution of sales across different locations.

```python
sale_by_outlet_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sale_by_outlet_location = sale_by_outlet_location.sort_values('Sales', ascending = False)

plt.figure(figsize=(8,3))
ax = sns.barplot(x = 'Sales', y = 'Outlet Location Type', data = sale_by_outlet_location, color='gold')

plt.title('Sales by Outlet Location')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout()
plt.show()
```
